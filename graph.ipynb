# Visualization
- visualization of the evaluated output from GPT using the files with the suffix `_evaluated`
1. zero-shot
2. advanced prompting
3. fine-tuning
4. Comparison of advanced prompting & zero-shot prompting
5. Distribution of advanced prompting & zero-shot prompting
6. Comparison of all approaches

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# source is the evaluated zero_shot outputs
file_path = 'analysis/zeroshot_evaluated.xlsx'
data = pd.read_excel(file_path)

# Setting the positions and width
positions = np.arange(len(data['company']))
width = 0.3

# Plot configuration
fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(positions, data['factuality'], width, label='Factuality', color='powderblue')
rects2 = ax.bar(positions, data['clarity'], width, bottom=data['factuality'], label='Clarity', color='lightcoral')
rects3 = ax.bar(positions, data['relevance'], width, bottom=data['factuality'] + data['clarity'], label='Relevance', color='wheat')
rects4 = ax.bar(positions, data['information Richness'], width, bottom=data['factuality'] + data['clarity'] + data['relevance'], label='Information Richness', color='darkseagreen')
rects5 = ax.bar(positions, data['on-pointedness'], width, bottom=data['factuality'] + data['clarity'] + data['relevance'] + data['information Richness'], label='On-pointedness', color='slategrey')

ax.set_xlabel('Company')
ax.set_ylabel('Scores')
ax.set_xticks(positions)
ax.set_xticklabels(data['company'], rotation=90)
ax.legend(loc='upper right', prop={'size': 8})

# Adding a legend and making layout tight
fig.tight_layout()


plt.show()
```

```python
#comparison fewshot & zeroshot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = 'analysis/fewshot_evaluated.xlsx'
data = pd.read_excel(file_path)

positions = np.arange(len(data['company']))
width = 0.3

fig, ax = plt.subplots(figsize=(15, 6))
rects1 = ax.bar(positions - width/2 , data['factuality_1'], width, label='Factuality', color='powderblue', alpha=0.6)
rects2 = ax.bar(positions- width/2, data['clarity_1'], width, bottom=1, label='Clarity', color='lightcoral', alpha=0.6)
rects3 = ax.bar(positions- width/2, data['relevance_1'], width, bottom=2, label='Relevance', color='wheat', alpha=0.7)
rects4 = ax.bar(positions- width/2, data['information_Richness_1'], width, bottom=3, label='Information Richness', color='darkseagreen', alpha=0.6)
rects5 = ax.bar(positions- width/2, data['on-pointedness_1'], width, bottom=4, label='On-pointedness', color='slategrey', alpha=0.6)
rects1 = ax.bar(positions+ 0.7*width, data['factuality'], width, color='powderblue')
rects2 = ax.bar(positions+ 0.7*width, data['clarity'], width, bottom=1,  color='lightcoral')
rects3 = ax.bar(positions+ 0.7*width, data['relevance'], width, bottom=2, color='wheat')
rects4 = ax.bar(positions+ 0.7*width, data['information_Richness'], width, bottom=3,  color='darkseagreen')
rects5 = ax.bar(positions+ 0.7*width, data['on-pointedness'], width, bottom= 4, color='slategrey')

ax.set_xlabel('Company')
ax.set_ylabel('Scores')

ax.set_xticks(positions)
ax.set_xticklabels(data['company'], rotation=90)

ax.legend(loc='right', bbox_to_anchor=(-0.015, 0.9),prop={'size': 7})


fig.tight_layout()

plt.show()
```

```python
#distribution fewshot and zeroshot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = 'analysis/fewshot_evaluated.xlsx'
data = pd.read_excel(file_path)

# Drop NaN values
few_shot_data = data['Few_shot'].dropna()
zero_shot_data = data['zero_shot'].dropna()

# Define the bins
bins = np.linspace(1, 6, 6)

# Calculate the histogram data
few_shot_hist, _ = np.histogram(few_shot_data, bins=bins)
zero_shot_hist, _ = np.histogram(zero_shot_data, bins=bins)

# Define the positions of the bins
bin_centers = 0.5 * (bins[:-1] + bins[1:])

plt.figure(figsize=(12, 6))


plt.plot(bin_centers, few_shot_hist, color='blue', marker='o', label='Few-Shot')
plt.plot(bin_centers, zero_shot_hist, color='red', marker='o', label='Zero-Shot')


plt.xlabel('Scores')
plt.ylabel('Amount of Companies')

plt.xticks(bin_centers, labels=[f'{int(b)}' for b in bins[:-1]])
plt.legend()


plt.show()

```

```python
#comparison all approaches
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = 'analysis/Overview.xlsx'  
data = pd.read_excel(file_path)

data_dropped = data.dropna()


sorted_data = data_dropped.sort_values(by=['Few_shot', 'zero_shot', 'finetuned_zero', 'finetuned_few'], ascending=False).reset_index(drop=True)


positions = np.arange(len(sorted_data))
width = 0.15


fig, ax = plt.subplots(figsize=(16, 6))


rects2 = ax.bar(positions - 1.5*width, sorted_data['zero_shot'], width, label='Zero Shot', color='powderblue')
rects1 = ax.bar(positions - 0.5*width, sorted_data['Few_shot'], width, label='Few Shot', color='lightcoral')
rects3 = ax.bar(positions + 0.7*width, sorted_data['finetuned_zero'], width, label='Zero Shot Fine-tuned', color='dodgerblue')
rects4 = ax.bar(positions + 1.7*width, sorted_data['finetuned_few'], width, label='Few-shot Fine-tuned', color='indianred')

ax.set_xlabel('Company')
ax.set_ylabel('Scores')

ax.set_xticks(positions)
ax.set_xticklabels(sorted_data['company'], rotation=90)
ax.legend(loc='upper right', prop={'size': 8})
fig.tight_layout()

plt.show()
```
